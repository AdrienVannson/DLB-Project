```python
import wandb
import pandas as pd
import json

wandb.login()
api = wandb.Api()
```

```python
runs = api.runs("epfl-dlb")
```

```python
results = []

for run in runs:
    if run.group != "relationnet_sp_5w_1s_mse_batchnorm":
        continue

    path = run.summary.eval_results.path
    run.file(path).download(replace=True)

    with open(path, 'r') as f:
        data = json.load(f)

        val = float(data["data"][1][2])
        test = float(data["data"][2][2])
    

    results.append({"name": run.name, "val": val, "test": test})

df = pd.DataFrame(results)
df.sort_values("val")
```
